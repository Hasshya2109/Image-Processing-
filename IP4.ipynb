```python
#Hasshya Shah
#C082
#To write a program to enhance the quality of an image (on your own photograph) by noise removal.
import numpy as np
import cv2
import random
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('pic2.jpg')
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resized_image = cv2.resize(image, (256, 256))
cv2_imshow(resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
for i in range(0,100):
    x = random.randint(0,255)
    y = random.randint(0,255)
    gray_image[x, y] = 255
cv2_imshow(gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel = np.array([[-1.0, -1.0, -1.0],
                   [-1.0, 8.0, -1.0],
                   [-1.0, -1.0, -1.0]])
hpf_image =   kernel/(np.sum(kernel) if np.sum(kernel)!=0 else 1)

kernel_lpf = np.ones((3, 3), np.float32) / 9
lpf_image = cv2.filter2D(gray_image, -3, kernel_lpf)

median_filtered_image = cv2.medianBlur(gray_image, 3)

hpf_image = cv2.filter2D(gray_image,-1,kernel)

# Display the original and filtered images
print("Image after applying High Pass Filter with center of kernal 4:")
cv2_imshow(hpf_image)
print("\nImage after applying Low Pass Filter:")
cv2_imshow(lpf_image)
print("\nImage after applying Median Filter:")
cv2_imshow(median_filtered_image)
```

```python
min_filtered=gray_image.copy()

for i in range(1,(gray_image.shape[0])-1):

    for j in range(1,(gray_image.shape[1])-1):

        neighbors=[gray_image[i-1][j-1],gray_image[i-1][j],gray_image[i-1][j+1],

                   gray_image[i][j-1],gray_image[i][j],gray_image[i][j+1],

                   gray_image[i+1][j-1],gray_image[i+1][j],gray_image[i+1][j+1]]

        min_filtered[i][j]=min(neighbors)


print("Image after applying min filter:")
cv2_imshow(min_filtered)

cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
max_filtered=gray_image.copy()

for i in range(1,(gray_image.shape[0])-1):

    for j in range(1,(gray_image.shape[1])-1):

        neighbors=[gray_image[i-1][j-1],gray_image[i-1][j],gray_image[i-1][j+1],

                   gray_image[i][j-1],gray_image[i][j],gray_image[i][j+1],

                   gray_image[i+1][j-1],gray_image[i+1][j],gray_image[i+1][j+1]]

        max_filtered[i][j]=max(neighbors)


print("Image after applying max filter:")
cv2_imshow( max_filtered)

cv2.waitKey(0)

cv2.destroyAllWindows()
```
