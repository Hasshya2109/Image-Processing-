```python
cv2_imshow(gray_img)
```

```python
def get_hadamard_matrix():
  h2 = np.array([[1,1],[1,-1]])
  new_h = h2
  for i in range(8):
      new_h = np.kron(new_h,h2)
  n_factor = 1/math.sqrt(new_h.shape[0])
  new_h = n_factor*new_h

  h_transpose = new_h.T

  return new_h,h_transpose
```

```python
new_h,h_transpose = get_hadamard_matrix()
```

# Applying the Hadamard Transform

**APPLY THE HADAMARD**

```python
h_transform_img = np.matmul(new_h,gray_img)
# apply hadamrd transpose
h_img = np.matmul(h_transform_img,h_transpose)
cv2_imshow(gray_img)
cv2_imshow(h_img)
```

**APPLY THE HADAMARD INVERSE**

```python
h_transform_img = np.matmul(new_h,h_img)
# apply hadamrd transpose
final_img = np.matmul(h_transform_img,h_transpose)
# cv2_imshow(h_transform_img)
cv2_imshow(final_img)
```

```python
def had2walsh(h4):
  h_dict = {}
  for i,row in enumerate(h4):
    count = 0
    for el in range(len(row)-1):
      if row[el] == row[el+1]:
        pass
      else:
        count+=1
      h_dict[i] = count

  sorted_welsh = sorted(h_dict.items(), key=lambda x:x[1])
  w4 = np.zeros(h4.shape)
  for k,v in sorted_welsh:
    w4[v] = h4[k]
  return w4,w4.T
```
